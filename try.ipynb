```python

from ultralytics import YOLO
import cv2
```

```python
model = YOLO("./detector/YOLOv8/yolov8s.pt")
video_path = "./001.avi"
```

```python
video_capture = cv2.VideoCapture(video_path)
```

```python
while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    results = model.predict(frame, classes=2)
    for result in results:
        detection = []
        class_name = []
        print(result.boxes.data.tolist())
        for r in result.boxes.data.tolist():
            print(r)
            x1, y1, x2, y2, score, class_id = r
            x1 = int(x1)
            x2 = int(x2)
            y1 = int(y1)
            y2 = int(y2)
            class_id = int(class_id)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            # cv2.putText(frame, str(score), (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            # cv2.putText(frame, str(class_id), (x1, y2), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            cv2.imshow("frame", frame)
        
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
cv2.destroyAllWindows()
video_capture.release()
```
